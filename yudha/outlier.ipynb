```python
from sklearn.decomposition import PCA

# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)  # You can adjust the number of components as needed
pca_result = pca.fit_transform(df_scaled)

# Add PCA results to the dataframe
df_scaled['PCA1'] = pca_result[:, 0]
df_scaled['PCA2'] = pca_result[:, 1]

# Clustering with PCA components
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca.fit(df_scaled[['PCA1', 'PCA2']])

# Adding cluster labels to the original dataframe
df_scaled['pca_cluster'] = kmeans_pca.labels_

# Visualizing the clusters with PCA
plt.figure(figsize=(10, 6))
for cluster in range(3):
    cluster_data = df_scaled[df_scaled['pca_cluster'] == cluster]
    plt.scatter(cluster_data['PCA1'], cluster_data['PCA2'], label=f'Cluster {cluster}')

plt.scatter(kmeans_pca.cluster_centers_[:, 0], kmeans_pca.cluster_centers_[:, 1], s=300, c='red', marker='*', label='Centroids')
plt.title('Clusters of Air Pollution Data (with PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()

# Displaying the counts of data points in each cluster with PCA
print("Counts of data points in each cluster (with PCA):")
print(df_scaled['pca_cluster'].value_counts())


# Saving the clustered data with PCA components to an Excel file
# df_scaled.to_excel("clustered_with_PCA.xlsx", index=False)
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Memilih fitur-fitur yang akan digunakan untuk clustering
features = ['co_aqi_value\t', 'ozone_aqi_value', 'no2_aqi_value', 'pm2.5_aqi_value']
X = df[features]

# Preprocessing data (scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Menentukan jumlah cluster
n_clusters = 4

# Inisialisasi dan fitting model KMeans
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X_scaled)

# Menambahkan label cluster ke dalam DataFrame
df['cluster_label'] = kmeans.labels_

# Visualisasi hasil clustering
sns.pairplot(df, hue='cluster_label', vars=features, palette='Set1')
plt.title('Clustering Results')
plt.show()

# Statistik rata-rata setiap cluster untuk fitur-fitur yang digunakan
cluster_stats = df.groupby('cluster_label')[features].mean()
print(cluster_stats)

# Visualizing the pairplot
pairplot = sns.pairplot(df, hue='cluster_label', vars=features, palette='Set1')
pairplot.fig.suptitle('Clustering Results', y=1.02)  # Adjust title position
plt.show()

# Saving the pairplot visualization to a file
pairplot.savefig("pairplot_clustering_results.png")

# Saving the DataFrame with cluster labels to an Excel file
# df.to_excel("clustered_data_with_pairplot.xlsx", index=False)

```
