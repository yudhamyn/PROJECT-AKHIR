```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Memilih fitur-fitur yang akan digunakan untuk clustering
features = ['co_aqi_value\t', 'ozone_aqi_value', 'no2_aqi_value', 'pm2.5_aqi_value']
X = df[features]

# Preprocessing data (scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Menentukan jumlah cluster
n_clusters = 2

# Inisialisasi dan fitting model KMeans
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X_scaled)

# Menambahkan label cluster ke dalam DataFrame
df['cluster_label'] = kmeans.labels_

# Visualisasi hasil clustering
sns.pairplot(df, hue='cluster_label', vars=features, palette='Set1')
plt.title('Clustering Results')
plt.show()

# Statistik rata-rata setiap cluster untuk fitur-fitur yang digunakan
cluster_stats = df.groupby('cluster_label')[features].mean()
print(cluster_stats)

```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# List untuk menyimpan nilai inersia (inertia) dari setiap iterasi
inertia = []

# Coba jumlah cluster dari 1 hingga 10 dan hitung inersia untuk setiap iterasi
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Method
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# List untuk menyimpan nilai silhouette score dari setiap iterasi
silhouette_scores = []

# Coba jumlah cluster dari 2 hingga 10 dan hitung silhouette score untuk setiap iterasi
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot Silhouette Score
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score')
plt.show()
```

```python
# Pemilihan jumlah cluster optimal berdasarkan metode Elbow
optimal_clusters_elbow = 3  # Misalnya, berdasarkan grafik Elbow Method, titik "siku" berada di sekitar 3 cluster

# Pemilihan jumlah cluster optimal berdasarkan Silhouette Score
optimal_clusters_silhouette = silhouette_scores.index(max(silhouette_scores)) + 2  # Ditambah 2 karena iterasi dimulai dari 2

print(f"Jumlah cluster optimal berdasarkan Elbow Method: {optimal_clusters_elbow}")
print(f"Jumlah cluster optimal berdasarkan Silhouette Score: {optimal_clusters_silhouette}")
```

```python
from sklearn.metrics import davies_bouldin_score

# Menghitung DBI untuk clustering dengan jumlah cluster optimal dari Elbow Method
kmeans_elbow = KMeans(n_clusters=optimal_clusters_elbow, random_state=42)
kmeans_elbow.fit(X_scaled)
dbi_elbow = davies_bouldin_score(X_scaled, kmeans_elbow.labels_)

# Menghitung DBI untuk clustering dengan jumlah cluster optimal dari Silhouette Score
kmeans_silhouette = KMeans(n_clusters=optimal_clusters_silhouette, random_state=42)
kmeans_silhouette.fit(X_scaled)
dbi_silhouette = davies_bouldin_score(X_scaled, kmeans_silhouette.labels_)

print(f"DBI untuk clustering dengan jumlah cluster optimal dari Elbow Method: {dbi_elbow}")
print(f"DBI untuk clustering dengan jumlah cluster optimal dari Silhouette Score: {dbi_silhouette}")
```
