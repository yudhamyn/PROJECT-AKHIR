```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df['PM2.5 AQI Category'].value_counts().plot.bar()
```

```python
col=[ 'AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']
fig = plt.figure(figsize=(20 ,5))
for i in range(len(col)):
    sns.histplot(df, x=col[i], ax=plt.subplot(1, 5, i+1))
    plt.title(col[i])
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))

for i in range(len(col)):
    plt.subplot(2,3,i+1)
    plt.title(col[i])
    sns.boxplot(data=df,x=df[col[i]],y=df['AQI Category'])
plt.tight_layout()
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```
